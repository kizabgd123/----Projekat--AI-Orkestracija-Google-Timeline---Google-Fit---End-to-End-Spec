```python
# High score (faktori koji POVEĆAVAJU potrebu za navodnjavanjem):
#   +2  ako Soil_Moisture < 25
#   +2  ako Rainfall_mm < 300
#   +1  ako Temperature_C > 30
#   +1  ako Wind_Speed_kmh > 10
#                               ─────────
#                      max High score = 6

# Low score (faktori koji SMANJUJU potrebu):
#   +2  ako Crop_Growth_Stage == "Harvest"
#   +2  ako Crop_Growth_Stage == "Sowing"
#   +1  ako Mulching_Used == "Yes"
#                               ─────────
#                      max Low score = 3

# Score = High score − Low score    (raspon: −3 do +6)
```

```python
import pandas as pd
import numpy as np
import os

# Provera postojanja fajlova pre učitavanja
if os.path.exists('train.csv') and os.path.exists('test.csv'):
    train = pd.read_csv('train.csv')
    test  = pd.read_csv('test.csv')
    print(f"✅ Podaci učitani: Train {train.shape}, Test {test.shape}")

    def deotte_formula(df):
        high_score = (
            2 * (df['Soil_Moisture'] < 25).astype(int) +
            2 * (df['Rainfall_mm'] < 300).astype(int) +
            1 * (df['Temperature_C'] > 30).astype(int) +
            1 * (df['Wind_Speed_kmh'] > 10).astype(int)
        )
        low_score = (
            2 * (df['Crop_Growth_Stage'] == 'Harvest').astype(int) +
            2 * (df['Crop_Growth_Stage'] == 'Sowing').astype(int) +
            1 * (df['Mulching_Used'] == 'Yes').astype(int)
        )
        score = high_score - low_score
        prediction = np.where(score <= 0, 'Low', np.where(score <= 3, 'Medium', 'High'))
        return prediction, score

    train_preds, train_scores = deotte_formula(train)
    from sklearn.metrics import balanced_accuracy_score
    ba = balanced_accuracy_score(train['Irrigation_Need'], train_preds)
    print(f"\n✅ Balanced Accuracy na TRAIN setu: {ba:.5f}")
else:
    print("❌ Fajlovi train.csv i test.csv nisu pronađeni. Molimo koristite ćeliju za upload na vrhu.")
```

```python
import numpy as np
import pandas as pd
import os

def deotte_exact_logistic(df):
    soil_lt_25  = (df['Soil_Moisture'] < 25).astype(float)
    temp_gt_30  = (df['Temperature_C'] > 30).astype(float)
    rain_lt_300 = (df['Rainfall_mm'] < 300).astype(float)
    wind_gt_10  = (df['Wind_Speed_kmh'] > 10).astype(float)
    flowering  = (df['Crop_Growth_Stage'] == 'Flowering').astype(float)
    harvest    = (df['Crop_Growth_Stage'] == 'Harvest').astype(float)
    sowing     = (df['Crop_Growth_Stage'] == 'Sowing').astype(float)
    vegetative = (df['Crop_Growth_Stage'] == 'Vegetative').astype(float)
    mulch_no   = (df['Mulching_Used'] == 'No').astype(float)
    mulch_yes  = (df['Mulching_Used'] == 'Yes').astype(float)

    logit_low = (16.3173 - 11.0237*soil_lt_25 - 5.8559*temp_gt_30 - 10.8500*rain_lt_300 - 5.8284*wind_gt_10 - 5.4155*flowering + 5.5073*harvest + 5.2299*sowing - 5.4617*vegetative)
    logit_medium = (4.6524 + 0.3290*soil_lt_25 - 0.0204*temp_gt_30 + 0.1542*rain_lt_300 + 0.0841*wind_gt_10 + 0.3586*flowering - 0.1348*harvest - 0.3547*sowing + 0.3334*vegetative + 0.1883*mulch_no + 0.0142*mulch_yes)
    logit_high = (-20.9697 + 10.6947*soil_lt_25 + 5.8763*temp_gt_30 + 10.6958*rain_lt_300 + 5.7444*wind_gt_10 + 5.0569*flowering - 5.3725*harvest - 4.8752*sowing + 5.1283*vegetative + 2.8131*mulch_no - 2.8755*mulch_yes)

    logits = np.column_stack([logit_low, logit_medium, logit_high])
    exp_logits = np.exp(logits - logits.max(axis=1, keepdims=True))
    probs = exp_logits / exp_logits.sum(axis=1, keepdims=True)
    classes = ['Low', 'Medium', 'High']
    return [classes[i] for i in probs.argmax(axis=1)], probs

# Load if missing and file exists
if 'test' not in globals() and os.path.exists('test.csv'):
    test = pd.read_csv('test.csv')

if 'test' in globals():
    test_preds_exact, test_probs = deotte_exact_logistic(test)
    submission_exact = pd.DataFrame({'id': test['id'], 'Irrigation_Need': test_preds_exact})
    submission_exact.to_csv('submission_exact_logistic.csv', index=False)
    print("✅ submission_exact_logistic.csv sačuvan!")
else:
    print("❌ 'test' varijabla nije dostupna. Molimo prvo učitajte test.csv.")
```

```python
# ═══════════════════════════════════════════════
# HYBRID: Formula predikcije + ML model ensemble
# Ovo je pristup koji postiže ~0.98200 (top 1)
# ═══════════════════════════════════════════════

import lightgbm as lgb
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import balanced_accuracy_score

# 1. Dodaj magic_score kao feature u ML model
def add_formula_features(df):
    df = df.copy()
    df['soil_lt_25']  = (df['Soil_Moisture'] < 25).astype(int)
    df['rain_lt_300'] = (df['Rainfall_mm'] < 300).astype(int)
    df['temp_gt_30']  = (df['Temperature_C'] > 30).astype(int)
    df['wind_gt_10']  = (df['Wind_Speed_kmh'] > 10).astype(int)

    df['high_score'] = (df['soil_lt_25']*2 + df['rain_lt_300']*2 +
                        df['temp_gt_30'] + df['wind_gt_10'])

    df['is_harvest'] = (df['Crop_Growth_Stage'] == 'Harvest').astype(int)
    df['is_sowing']  = (df['Crop_Growth_Stage'] == 'Sowing').astype(int)
    df['mulch_yes']  = (df['Mulching_Used'] == 'Yes').astype(int)

    df['low_score']  = df['is_harvest']*2 + df['is_sowing']*2 + df['mulch_yes']
    df['magic_score'] = df['high_score'] - df['low_score']

    # Formula predikcija kao feature
    df['formula_pred'] = np.where(
        df['magic_score'] <= 0, 0,
        np.where(df['magic_score'] <= 3, 1, 2)
    )
    return df

train_fe = add_formula_features(train)
test_fe  = add_formula_features(test)

# 2. Formula predikcije
formula_preds_test, _ = deotte_formula(test)

# 3. ML model sa formula features
feature_cols = [c for c in train_fe.columns
                if c not in ['id', 'Irrigation_Need']]

# Encode categorical columns correctly
cat_features = ['Crop_Type', 'Soil_Type', 'Region', 'Weather_Condition',
                'Crop_Growth_Stage', 'Mulching_Used', 'Irrigation_System_Type',
                'Season', 'Irrigation_Type', 'Water_Source']

for col in cat_features:
    if col in train_fe.columns:
        train_fe[col] = train_fe[col].astype('category')
        test_fe[col]  = test_fe[col].astype('category')

label_map = {'Low': 0, 'Medium': 1, 'High': 2}
y = train_fe['Irrigation_Need'].map(label_map)

X_train = train_fe[feature_cols]
X_test  = test_fe[feature_cols]

lgb_params = {
    'objective': 'multiclass',
    'num_class': 3,
    'metric': 'multi_logloss',
    'learning_rate': 0.05,
    'num_leaves': 63,
    'max_depth': 7,
    'subsample': 0.8,
    'colsample_bytree': 0.7,
    'verbose': -1,
    'n_jobs': -1
}

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
ml_probs_test = np.zeros((len(test), 3))
oof_preds = np.zeros(len(train))

for fold, (tr_idx, val_idx) in enumerate(skf.split(X_train, y)):
    dtrain = lgb.Dataset(X_train.iloc[tr_idx], label=y.iloc[tr_idx])
    dval   = lgb.Dataset(X_train.iloc[val_idx], label=y.iloc[val_idx])

    model = lgb.train(lgb_params, dtrain,
                      valid_sets=[dval],
                      num_boost_round=1000,
                      callbacks=[lgb.early_stopping(50), lgb.log_evaluation(200)])

    val_probs = model.predict(X_train.iloc[val_idx])
    oof_preds[val_idx] = val_probs.argmax(axis=1)
    ml_probs_test += model.predict(X_test) / 5

    fold_ba = balanced_accuracy_score(y.iloc[val_idx], val_probs.argmax(axis=1))
    print(f"  Fold {fold+1} BA: {fold_ba:.5f}")

oof_ba = balanced_accuracy_score(y, oof_preds)
print(f"\n📊 OOF Balanced Accuracy (ML): {oof_ba:.5f}")

# 4. HYBRID BLEND
ml_preds_labels = ['Low', 'Medium', 'High']
ml_final = [ml_preds_labels[i] for i in ml_probs_test.argmax(axis=1)]

_, test_scores = deotte_formula(test)
hybrid_preds = []

for i in range(len(test)):
    score = test_scores[i]
    if score <= -1 or score >= 5:
        hybrid_preds.append(formula_preds_test[i])
    else:
        hybrid_preds.append(ml_final[i])

submission_hybrid = pd.DataFrame({
    'id': test['id'],
    'Irrigation_Need': hybrid_preds
})
submission_hybrid.to_csv('submission_hybrid.csv', index=False)
print("\n✅ submission_hybrid.csv sačuvan!")
print(f"\nDistribucija:\n{submission_hybrid['Irrigation_Need'].value_counts()}")
```

### 🎯 Optimizacija pragova (Threshold Tuning)
Standardni `.argmax()` pretpostavlja prag od 0.33 za svaku klasu. Pošto je 'High' retka, njena verovatnoća će često biti potisnuta. Koristimo `scipy.optimize` da nađemo pragove koji maksimizuju Balanced Accuracy na OOF (Out-of-Fold) predikcijama.

```python
import numpy as np
from scipy.optimize import minimize
from sklearn.metrics import balanced_accuracy_score

# Pretpostavljamo da oof_probs već postoji iz prethodnog treniranja
# Ako nemamo oof_probs, uzećemo verovatnoće iz poslednjeg treninga
def objective(thresholds, y_true, y_probs):
    # Modifikujemo verovatnoće težinama
    weighted_probs = y_probs * thresholds
    preds = np.argmax(weighted_probs, axis=1)
    return -balanced_accuracy_score(y_true, preds)

# Početni pragovi su 1.0 za sve klase
initial_thresholds = [1.0, 1.0, 1.0]

# Optimizacija
# y je label-encoded (0, 1, 2), oof_preds_probs su verovatnoće modela
# Ovde koristimo model.predict(X_train) kao proxy ako nemamo sačuvan OOF
# Ali pravi Grandmaster uvek koristi OOF!
res = minimize(objective, initial_thresholds, args=(y, model.predict(X_train)),
               method='Nelder-Mead', tol=1e-6)

best_thresholds = res.x
print(f"Optimalni pragovi: {best_thresholds}")

# Primena na test set
final_test_probs = ml_probs_test * best_thresholds
final_preds = np.argmax(final_test_probs, axis=1)

# Mapiranje nazad u stringove
inv_label_map = {v: k for k, v in label_map.items()}
submission_tuned = pd.DataFrame({
    'id': test['id'],
    'Irrigation_Need': [inv_label_map[p] for p in final_preds]
})
submission_tuned.to_csv('submission_tuned.csv', index=False)
print("✅ submission_tuned.csv je generisan sa optimizovanim pragovima!")
```

```python
import pandas as pd
import numpy as np

def advanced_fe(df):
    df = df.copy()

    # 1. Klimatski indeksi
    df['Heat_Index'] = df['Temperature_C'] * (df['Humidity'] / 100)
    df['Moisture_Stress'] = df['Soil_Moisture'] / (df['Temperature_C'] + 1e-5)

    # 2. Interakcije sa padavinama
    df['Rain_per_Temp'] = df['Rainfall_mm'] / (df['Temperature_C'] + 1e-5)

    # 3. Grupne statistike (Deviation from mean)
    # Ovo pomaže modelu da vidi da li je polje 'suvše' u odnosu na svoj region/tip zemljišta
    for col in ['Soil_Type', 'Region']:
        df[f'{col}_Avg_Moisture'] = df.groupby(col)['Soil_Moisture'].transform('mean')
        df[f'{col}_Moisture_Diff'] = df['Soil_Moisture'] - df[f'{col}_Avg_Moisture']

    # 4. Binarni indikatori za ekstremne uslove
    df['Is_Dry_And_Hot'] = ((df['Soil_Moisture'] < 20) & (df['Temperature_C'] > 35)).astype(int)

    return df

train_advanced = advanced_fe(train)
test_advanced = advanced_fe(test)

print(f"Novi broj feature-a: {train_advanced.shape[1]}")
display(train_advanced.head(3))
```

```python
# ═══════════════════════════════════════════════
# RE-TRAINING SA NAPREDNIM FEATURE-IMA
# ═══════════════════════════════════════════════

# Ponovno definisanje feature kolona i kategorija
feature_cols_adv = [c for c in train_advanced.columns if c not in ['id', 'Irrigation_Need']]
cat_features_adv = ['Crop_Type', 'Soil_Type', 'Region', 'Weather_Condition',
                    'Crop_Growth_Stage', 'Mulching_Used', 'Irrigation_System_Type',
                    'Season', 'Irrigation_Type', 'Water_Source']

# Priprema podataka - Provera postojanja kolona
for col in cat_features_adv:
    if col in train_advanced.columns:
        train_advanced[col] = train_advanced[col].astype('category')
    if col in test_advanced.columns:
        test_advanced[col]  = test_advanced[col].astype('category')

y_adv = train_advanced['Irrigation_Need'].map(label_map)
X_adv = train_advanced[feature_cols_adv]

# 5-Fold CV sa novim podacima
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
oof_probs_adv = np.zeros((len(train_advanced), 3))
ml_probs_test_adv = np.zeros((len(test_advanced), 3))

for fold, (tr_idx, val_idx) in enumerate(skf.split(X_adv, y_adv)):
    dtrain = lgb.Dataset(X_adv.iloc[tr_idx], label=y_adv.iloc[tr_idx])
    dval   = lgb.Dataset(X_adv.iloc[val_idx], label=y_adv.iloc[val_idx])

    model_adv = lgb.train(lgb_params, dtrain,
                          valid_sets=[dval],
                          num_boost_round=1000,
                          callbacks=[lgb.early_stopping(50), lgb.log_evaluation(200)])

    oof_probs_adv[val_idx] = model_adv.predict(X_adv.iloc[val_idx])
    ml_probs_test_adv += model_adv.predict(test_advanced[feature_cols_adv]) / 5

    fold_ba = balanced_accuracy_score(y_adv.iloc[val_idx], oof_probs_adv[val_idx].argmax(axis=1))
    print(f"  Fold {fold+1} BA: {fold_ba:.5f}")

# Finalna optimizacija pragova na novom OOF-u
res_adv = minimize(objective, initial_thresholds, args=(y_adv, oof_probs_adv),
                   method='Nelder-Mead', tol=1e-6)

print(f"\nOptimalni pragovi (Napredni): {res_adv.x}")

# Generisanje finalnog submissiona
final_test_probs_adv = ml_probs_test_adv * res_adv.x
final_preds_adv = np.argmax(final_test_probs_adv, axis=1)

submission_adv = pd.DataFrame({
    'id': test_advanced['id'],
    'Irrigation_Need': [inv_label_map[p] for p in final_preds_adv]
})
submission_adv.to_csv('submission_advanced_tuned.csv', index=False)
print("\n✅ submission_advanced_tuned.csv je spreman za leaderboard!")
```

### 🏁 Finalni Ensemble & Leaderboard Strategija

Na osnovu analize takmičenja **S6E4**, ključni faktori za uspeh su:

1.  **Leaderboard Split (20/80)**: Javni poredak (Public LB) se zasniva na samo 20% test podataka. S obzirom na to da je ciljano rešenje blizu savršenstva (0.98219), stabilnost modela na preostalih 80% (Private LB) je kritična kako bi se izbegao 'shake-up'.
2.  **Deotte's Magic Formula**: Originalni dataset je generisan po strogoj matematičkoj logici. Iako ML modeli (LightGBM/XGBoost) odlično aproksimiraju ove zavisnosti, direktna primena formule na ekstremnim vrednostima (`score <= -2` ili `score >= 5`) eliminiše greške koje model može napraviti zbog šuma.
3.  **Optimizacija Pragova**: Pošto je metrika **Balanced Accuracy**, a klasa 'High' je izuzetno retka (~3.3%), standardni pragovi nisu dovoljni. Nelder-Mead optimizacija težina za svaku klasu omogućava nam da preciznije 'pogodimo' retke slučajeve bez narušavanja performansi na ostalim klasama.

Ovaj pristup nam omogućava da se takmičimo za sam vrh tabele, ciljajući rezultat od **0.98219**.

```python
import pandas as pd
import numpy as np

# 1. Uzimamo predikcije iz modela (koristimo submission_tuned koji je dostupan)
# Ako submission_adv ne postoji zbog prekida, koristimo prethodnu najbolju verziju
try:
    final_sub = submission_adv.copy()
    print("Korišćenje advanced tuned modela.")
except NameError:
    final_sub = submission_tuned.copy()
    print("Korišćenje standardnog tuned modela (fallback).")

# 2. Primena 'Magične Formule' kao korektivnog faktora
# Ako formula daje ekstremno visoke ili niske skorove, verujemo njoj više nego modelu
_, test_scores = deotte_formula(test)

# Logika: Formula je izuzetno precizna na ekstremima
for i in range(len(test)):
    score = test_scores[i]
    if score <= -2: # Siguran 'Low'
        final_sub.at[i, 'Irrigation_Need'] = 'Low'
    elif score >= 5: # Siguran 'High'
        final_sub.at[i, 'Irrigation_Need'] = 'High'

# 3. Čuvanje finalnog fajla
final_sub.to_csv('final_submission_0.982.csv', index=False)
print("\n✅ Finalni ensemble submission je spreman!")
print(final_sub['Irrigation_Need'].value_counts())
```

### 🔗 Korelacija Feature-a sa Ciljnom Promenljivom

### ⚖️ Regularizacija Meta-modela
Smanjujemo uticaj `magic_score` uvođenjem jače L2 kazne (`C=0.01`) kako bismo naterali model da balansira između formule i ML predikcija.

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import pandas as pd

# 1. Priprema podataka (isto kao ranije, koristimo OOF i test verovatnoće)
try:
    # Pokušavamo sa oof_probs_adv ako su dostupni, inače fallback na oof_preds
    if 'oof_probs_adv' in globals() and np.sum(oof_probs_adv) > 0:
        X_train_meta = np.column_stack([oof_probs_adv, train_fe['magic_score'].values])
        X_test_meta = np.column_stack([ml_probs_test_adv, test_fe['magic_score'].values])
    else:
        # Fallback na oof_preds iz prvog modela
        oof_one_hot = pd.get_dummies(oof_preds).values
        X_train_meta = np.column_stack([oof_one_hot, train_fe['magic_score'].values])
        X_test_meta = np.column_stack([ml_probs_test, test_fe['magic_score'].values])

    # 2. Trening sa JAKOM regularizacijom (C=0.01)
    # Manji C znači jača regularizacija
    reg_meta_model = LogisticRegression(C=0.01, multi_class='multinomial', max_iter=1000, penalty='l2')
    reg_meta_model.fit(X_train_meta, y)

    # 3. Analiza novih koeficijenata
    print("Novi koeficijenti (sa regularizacijom):")
    feat_names = ['Prob_0', 'Prob_1', 'Prob_2', 'Magic_Score']
    coef_df_reg = pd.DataFrame(reg_meta_model.coef_, columns=feat_names, index=['Low', 'Medium', 'High'])
    display(coef_df_reg)

    # 4. Predikcije i submission
    meta_preds_reg = reg_meta_model.predict(X_test_meta)
    submission_reg = pd.DataFrame({
        'id': test['id'],
        'Irrigation_Need': [inv_label_map[p] for p in meta_preds_reg]
    })

    submission_reg.to_csv('submission_stacking_regularized.csv', index=False)
    print("\n✅ submission_stacking_regularized.csv je generisan!")
    print(f"Distribucija:\n{submission_reg['Irrigation_Need'].value_counts()}")

except Exception as e:
    print(f"❌ Greška: {e}")
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Definisanje ekstremnih vrednosti
extreme_mask = (train_magic_scores <= -2) | (train_magic_scores >= 4)

# 2. Filtriranje podataka
y_true_extreme = target_numeric[extreme_mask]
y_pred_extreme = oof_preds[extreme_mask]

# 3. Računanje matrice konfuzije
cm = confusion_matrix(y_true_extreme, y_pred_extreme)
labels = ['Low', 'Medium', 'High']

# 4. Vizualizacija
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.title(f'Confusion Matrix za ekstremne Magic Score vrednosti (n={extreme_mask.sum()})')
plt.xlabel('Predviđeno (ML Model)')
plt.ylabel('Stvarno (Ground Truth)')
plt.show()

# Procentualni prikaz grešaka
print("Procentualna tačnost na ekstremima po klasama:")
for i, label in enumerate(labels):
    class_total = cm[i].sum()
    if class_total > 0:
        acc = cm[i, i] / class_total * 100
        print(f"{label}: {acc:.2f}%")
```
