### 🧠 The Logic Engine & Pattern Registry

Now that we have the source code and the pattern data, let's look at how the **Self-Healer** identifies specific issues:

*   **`friction_patterns.json`**: Acts as the system's 'memory' or 'knowledge base', containing known error signatures.
*   **`diagnosis_engine.py`**: The active 'brain' that scans logs or outputs, compares them against the registry, and triggers the fix.

```python
import re

# Searching for playbook execution logic in the engine code
playbook_execution_pattern = r'def\s+(execute_playbook|run_recovery).*?:'
matches = re.finditer(playbook_execution_pattern, engine_code, re.DOTALL)

print("--- Playbook Execution Logic Discovery ---")
for match in matches:
    start = match.start()
    # Capture a significant chunk of the function logic
    print(engine_code[start:start+1000] + "...\n")

# Also let's look at the structure of a single pattern's playbook
sample_pattern = patterns[0]
print("Sample Playbook Structure:")
print(json.dumps(sample_pattern.get('playbook', []), indent=2))
```

### ⚙️ The Execution Blueprint

The `recovery_playbook` isn't just a list of commands; it's a tiered response strategy. Based on the code inspection, here is how the engine handles the recovery steps:

1.  **Step-by-Step Execution**: The engine iterates through the `playbook` array.
2.  **Command Types**: It supports `shell` commands (via `subprocess`) and internal `python` calls.
3.  **Conditional Continuation**: If a step fails, the engine decides whether to attempt the next step or trigger an immediate **Deadlock** escalation.
4.  **Logging**: Every attempt and its result are logged for the operator to review if needed.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Extracting pattern data for visualization
patterns = patterns_data.get('patterns', [])
data = []
for p in patterns:
    data.append({
        'Category': p.get('category', 'N/A'),
        'Pattern': p.get('regex_pattern', '')[:30] + '...',
        'Action': p.get('recovery_playbook', 'N/A')
    })

df = pd.DataFrame(data)

# Display as a table
print("--- Pattern Registry Overview ---")
display(df)

# Plot category distribution
df['Category'].value_counts().plot(kind='barh', color='lightgreen', title='Frequency of Error Categories in Registry')
plt.xlabel('Count')
plt.show()
```

### 🔄 The Automated Recovery Loop

Based on the code, the engine follows this strict workflow:
1. **Input**: Receives a raw error string (e.g., from a log file).
2. **Match**: Iterates through the `friction_patterns.json` using regex.
3. **Execute**: If a match is found, it triggers the associated shell command or python function defined in the `recovery_playbook`.
4. **Verify**: It checks if the error persists. If it does, it increments a retry counter or escalates to a 'Deadlock' state.

### 📊 Visual Architecture
Below is a visualization of the Self-Healer's file structure and operational flow.

```python
import matplotlib.pyplot as plt
import networkx as nx

def draw_architecture():
    G = nx.DiGraph()
    nodes = [
        ('SKILL.md', 'Instruction\nCenter'),
        ('friction_patterns.json', 'Pattern\nRegistry'),
        ('diagnosis_engine.py', 'Logic\nEngine'),
        ('Operator', 'Human\n(Deadlock Only)')
    ]
    for node, label in nodes:
        G.add_node(node, label=label)

    G.add_edge('SKILL.md', 'diagnosis_engine.py')
    G.add_edge('friction_patterns.json', 'diagnosis_engine.py')
    G.add_edge('diagnosis_engine.py', 'Operator', label='Deadlock')

    pos = nx.spring_layout(G)
    plt.figure(figsize=(10, 6))

    nx.draw(G, pos, with_labels=False, node_size=3000, node_color='skyblue', font_size=10, font_weight='bold', edge_color='gray')

    labels = nx.get_node_attributes(G, 'label')
    nx.draw_networkx_labels(G, pos, labels)

    plt.title("Self-Healer Architecture Flow")
    plt.show()

draw_architecture()
```
